```python
df = pd.read_csv('wine-clustering.csv')
print(df)
```

```python
df = df[['Alcohol','Malic_Acid','Ash','Ash_Alcanity','Magnesium','Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols', 'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline' ]]
print(df)
```

```python
df = df.fillna(df.mean()) # fill any empty cell with the meean value
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

km =KMeans(n_clusters =5)
km.fit(df)
df['cluster'] = km.labels_# add new column cluster to the dataset and assign the label to each row
print(df) # the data set has 1 new column cluste
```

```python

df1 = df[df.cluster ==0] # assign row with cluster=0 or df[df[‘cluster’] = 0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python

plt.scatter(df1.Alcohol, df1.Ash_Alcanity, color ='green')
plt.scatter(df2.Alcohol, df2.Ash_Alcanity, color ='red')
plt.scatter(df3.Alcohol, df3.Ash_Alcanity, color ='blue')
plt.xlabel('Alcohol')
plt.ylabel('Ash_Alcanity')
plt.scatter(km.cluster_centers_[ : ,0], km.cluster_centers_[:,3], color='black', marker ='*')
plt.show()
```

```python
plt.scatter(km.cluster_centers_[ : ,0], km.cluster_centers_[:,3], color='black', marker ='*')
plt.show()
```

```python
sse = []
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, n_init= 10, max_iter = 200)
  kmeans.fit(df)
  sse.append(kmeans.inertia_)
```

```python
plt.style.use("Solarize_Light2")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```
